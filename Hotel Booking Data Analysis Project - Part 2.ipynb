### T1_  1. What is the distribution of the number and types of special requests made by guests?

```python
ax = sns.histplot(df['total_of_special_requests'], binwidth = 1, color = 'tab:green')
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.title('Distribution of Special Requests')
plt.xlabel('Special Requests Made')
plt.ylabel('Number of Guests')
plt.show()
```

### T2_1. Which marketing channels and market segments contribute the most to successful bookings?

```python
mkt_succesful_bookings = df.loc[:, ['market_segment', 'distribution_channel']][df['is_canceled'] == 0]
```

```python
mkt_seg_count = mkt_succesful_bookings.iloc[:, 0].value_counts().sort_values(ascending = True)
mkt_channel_count = mkt_succesful_bookings.iloc[:, 1].value_counts().sort_values(ascending = True)
```

```python
seg_count = mkt_seg_count.to_list()
channel_count = mkt_channel_count.to_list()
```

```python
fig, ax = plt.subplots(1, 2, figsize = (20, 7))


bar_container = ax[0].barh(y = mkt_channel_count.index,
       width = mkt_channel_count, 
         color = ["xkcd:sky blue", 'xkcd:bubblegum', 'xkcd:pinky red', 'xkcd:marine'])
ax[0].bar_label(container = bar_container, 
             fmt = '{:,.0f}')

ax[0].set(xlabel = 'Number of Succesfull Bookings', 
          ylabel = 'Marketing Channels',
         title = 'Top Marketing Channels Contributing to Successful Booking')


bar_container = ax[1].barh(y = mkt_seg_count.index,
       width = mkt_seg_count, 
         color = ["xkcd:sky blue", 'xkcd:bubblegum', 'xkcd:pinky red', 'xkcd:marine'])
ax[1].bar_label(container = bar_container, 
             fmt = '{:,.0f}')
ax[1].set(xlabel = 'Number of Succesfull Bookings', 
          ylabel = 'Marketing Segments',
         title = 'Top Marketing Segments Contributing to Successful Booking')
# ax[1].set_title('Top Marketing Segments Contributing to Successful Booking')
# ax[1].set_xlabel('Number of Succesfull Bookings')
# ax[1].set_ylabel('Marketing Segments')

plt.tight_layout()
```

### T2_3. What is the distribution of bookings across various distribution channels?

```python
df = pd.read_csv("DS1_C5_S5_Hotel Dataset.csv")
```

```python
import matplotlib.pyplot as plt

# Count and sort the distribution channels
mkt_channel_count = df['distribution_channel'].value_counts().sort_values(ascending=False)

# Plot using Matplotlib
plt.figure(figsize=(8, 5))
bar_container = plt.bar(mkt_channel_count.index, mkt_channel_count.values)

# Add data labels to the bars
plt.bar_label(container = bar_container, 
             fmt = '{:,.0f}')

#for i, value in enumerate(mkt_channel_count.values):
#    plt.text(i, value, str(value), ha='center', va='bottom', fontsize=10)

# Add titles and labels
plt.title('Distribution of Bookings Across Various Distribution Channels', fontsize=16)
plt.xlabel('Distribution Channel', fontsize=14)
plt.ylabel('Number of Bookings', fontsize=14)

# Show the plot
plt.show()
```
